```python
selected_columns = selected_columns.applymap(
    lambda x: x.lower() if isinstance(x, str) else x
)

selected_columns['lokasi'] = selected_columns['lokasi'].str.strip()

# Tambahkan spasi setelah kec./kel./ds. jika belum ada
selected_columns['lokasi'] = selected_columns['lokasi'].str.replace(r'\b(kec|kel|ds|kp)\.(?!\s)', r'\1. ', regex=True)

selected_columns['lokasi'] = selected_columns['lokasi'].str.replace(r'\bkp\.', 'kampung', regex=True)
selected_columns['lokasi'] = selected_columns['lokasi'].str.replace(r'\b(kec|kel|ds)\.', r'\1', regex=True)


selected_columns.head()
```

```python
import pandas as pd

selected_columns['tanggal'] = pd.to_datetime(selected_columns['tanggal'], errors='coerce')

selected_columns['bulan'] = selected_columns['tanggal'].dt.month
selected_columns = selected_columns.drop(columns=['tanggal'])

selected_columns['tahun'] = selected_columns['tahun'].astype(int)
selected_columns['bulan'] = selected_columns['bulan'].astype(int)

selected_columns.head()
```

```python
rows = []

# Buffer baris kecamatan
last_kecamatan_row = None
kelurahan_found = False

for i in range(len(selected_columns)):
    raw_lokasi = selected_columns.loc[i, 'lokasi']
    lokasi = str(raw_lokasi).strip().lower() if pd.notnull(raw_lokasi) and str(raw_lokasi).strip() != '' else None

    if lokasi is None:
        # Lokasi kosong â†’ tetap simpan dengan None semua
        row_data = selected_columns.loc[i].to_dict()
        row_data['kecamatan'] = None
        row_data['kelurahan'] = None
        rows.append(row_data)

    elif lokasi.startswith('kec') or not lokasi.startswith(('kel', 'ds', 'desa', 'kampung')):
        # Simpan kecamatan sebelumnya kalau belum ada kelurahannya
        if last_kecamatan_row and not kelurahan_found:
            rows.append({
                **last_kecamatan_row,
                'kecamatan': last_kecamatan_row['lokasi'],
                'kelurahan': None
            })
        # Update buffer
        last_kecamatan_row = selected_columns.loc[i].to_dict()
        kelurahan_found = False

    elif lokasi.startswith(('kel', 'ds', 'desa', 'kampung')):
        if last_kecamatan_row:
            rows.append({
                **last_kecamatan_row,
                'kecamatan': last_kecamatan_row['lokasi'],
                'kelurahan': selected_columns.loc[i, 'lokasi']
            })
            kelurahan_found = True

# Simpan kecamatan terakhir kalau belum punya kelurahan
if last_kecamatan_row and not kelurahan_found:
    rows.append({
        **last_kecamatan_row,
        'kecamatan': last_kecamatan_row['lokasi'],
        'kelurahan': None
    })

# Buat DataFrame akhir
df_clean = pd.DataFrame(rows).drop(columns=['lokasi']).reset_index(drop=True)

df_clean['kecamatan'] = df_clean['kecamatan'].str.replace(r'\b(kec )\s?', '', regex=True)
df_clean['kelurahan'] = df_clean['kelurahan'].str.replace(r'\b(kel |ds |desa )\s?', '', regex=True)
```

```python
import os
import json
import pandas as pd

# Paths
file_31 = 'data-indonesia-master/kabupaten/31.json'
file_32 = 'data-indonesia-master/kabupaten/32.json'
file_36 = 'data-indonesia-master/kabupaten/36.json'

kecamatan_folder = 'data-indonesia-master/kecamatan/'

# Target nama kab/kota dari 32.json
target_jabar = {
    "KOTA BOGOR", "KAB. BOGOR", "KOTA DEPOK",
    "KOTA TANGERANG", "KOTA TANGERANG SELATAN", "KAB. TANGERANG",
    "KOTA BEKASI", "KAB. BEKASI"
}

target_banten = {
    "KOTA TANGERANG", "KOTA TANGERANG SELATAN", "KAB. TANGERANG"
}

rows = []

# Function untuk ambil data kecamatan dari satu file kabupaten
def proses_kabupaten(kabupaten_data, target_nama=None):
    for item in kabupaten_data:
        kab_id = item['id']
        kab_nama_upper = item['nama'].upper()
        kab_nama = item['nama'].title()

        # Jika ada target_nama, filter berdasarkan itu
        if target_nama and kab_nama_upper not in target_nama:
            continue

        kecamatan_file = os.path.join(kecamatan_folder, f"{kab_id}.json")

        if os.path.exists(kecamatan_file):
            with open(kecamatan_file, 'r', encoding='utf-8') as f:
                kecamatan_data = json.load(f)

            for kec in kecamatan_data:
                lat = kec.get('latitude')
                long = kec.get('longitude')

                rows.append({
                    'kabkot': kab_nama,
                    'kecamatan': kec['nama'].title(),
                    'lat': lat,
                    'long': long
                })
        else:
            print(f"File kecamatan tidak ditemukan untuk ID: {kab_id}")


# DKI Jakarta â†’ semua
with open(file_31, 'r', encoding='utf-8') as f:
    kabupaten_jkt = json.load(f)
proses_kabupaten(kabupaten_jkt)

# Jawa Barat â†’ pakai target_jabar
with open(file_32, 'r', encoding='utf-8') as f:
    kabupaten_jabar = json.load(f)
proses_kabupaten(kabupaten_jabar, target_nama=target_jabar)

# Banten â†’ pakai target_banten
with open(file_36, 'r', encoding='utf-8') as f:
    kabupaten_banten = json.load(f)
proses_kabupaten(kabupaten_banten, target_nama=target_banten)
```

```python
# Imputasi berdasarkan rata-rata suhu pada kabupaten, bulan, dan tahun yang sama
df['avg_temperature'] = df.groupby(['NAME_2', 'month', 'year'])['avg_temperature'].transform(
    lambda x: x.fillna(x.mean())
)

# Jika masih ada nilai kosong, gunakan rata-rata kabupaten dan bulan (tanpa mempertimbangkan tahun)
if df['avg_temperature'].isna().any():
    df['avg_temperature'] = df.groupby(['NAME_2', 'month'])['avg_temperature'].transform(
        lambda x: x.fillna(x.mean())
    )

# Jika masih ada nilai kosong, gunakan rata-rata kabupaten saja
if df['avg_temperature'].isna().any():
    df['avg_temperature'] = df.groupby(['NAME_2'])['avg_temperature'].transform(
        lambda x: x.fillna(x.mean())
    )

# Jika masih ada yang kosong (kabupaten tanpa data sama sekali), gunakan lat/long
if df['avg_temperature'].isna().any():
    # Identifikasi baris yang masih memiliki nilai kosong
    missing_idx = df['avg_temperature'].isna()
    
    # Untuk setiap baris dengan nilai kosong
    for idx in df[missing_idx].index:
        # Ambil lat/long dari baris tersebut
        lat_target = df.loc[idx, 'lat']
        long_target = df.loc[idx, 'long']
        
        # Hitung jarak ke setiap baris dengan nilai avg_temperature tidak kosong
        df_valid = df[~df['avg_temperature'].isna()]
        
        # Jarak Euclidean sederhana
        df_valid['distance'] = ((df_valid['lat'] - lat_target)**2 + 
                               (df_valid['long'] - long_target)**2)**0.5
        
        # Ambil k tetangga terdekat
        neighbors = df_valid.nsmallest(3, 'distance')
        
        # Imputasi dengan rata-rata tertimbang jarak
        df.loc[idx, 'avg_temperature'] = neighbors['avg_temperature'].mean()
```
